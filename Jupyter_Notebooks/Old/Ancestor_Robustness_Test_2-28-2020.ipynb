# aevol_misc_ancestor_robustness Test
In this notebook I will run a quick analysis of output from the post-treatment *aevol_misc_ancestor_robustness*. The data for this notebook is probably invalid but the coding framework is the important part.

```python
# Set up the environment

import pandas as pd
import matplotlib.pyplot as plt

INPUT_FILE = "C:\\Users\\davis\\Desktop\\ancestor_robustness_best.out"
OUTPUT_DIR = "C:\\Users\\davis\\Desktop\\robustness_output\\"

robustness_names = ['generation', 'fract_positive_offspring', 'fract_neutral_offspring', 'fract_neutral_mutants', 'fract_negative_offspring', 'cumul_delt-gaps_positive_offspring', 'cumul_delt-gaps_negative_offspring', 'delta-gap_best_offspring', 'delta-gap_worst_offspring', 'cumul_delta-fitness_positive_offspring', 'cumul_delta-fitness_negative_offspring', 'delta-fitness_best_offspring', 'delta-fitness_worst_offspring']
print("Reading in data...")
df_robustness = pd.read_csv(INPUT_FILE, skiprows=19, delim_whitespace=True, header=0, names=robustness_names)
print("done.")
```

## Graph the results

```python
fig_list = []

# Loop over all columns and graph them by generation
for label in robustness_names:
    if(label == 'generation'):
        continue
    else:
        # Create new graph
        fig_robustness = plt.figure(figsize=(30,20))
        ax_robustness = fig_robustness.add_subplot(111)
        ax_robustness.set_title(label, fontsize=32)
        ax_robustness.set_xlabel("Generation", fontsize=20)
        ax_robustness.set_ylabel(label, fontsize=20)
        df_robustness.plot(x='generation', y=label, label=label, ax=ax_robustness)
        fig_robustness.savefig(OUTPUT_DIR + "robustness_" + label+".png")
        
```
