## Comparing Genome Length

We should compare the length of the genomes for all seven conditions and see if changing the parameters had an effect on the length of the genome. 
Following the method of the paper "How to Reduce a Genome?", we compute the average length across all seeds.

##### We begin with the Control condition.

```python
# Seed01
f_c_1 = open(INPUT_ROOT_DIR + "seed01//control//best_last_org.txt", "r")
txt_c_1 = f_c_1.read()
size_c_1 = len(txt_c_1)
f_c_1.close()

# Seed02
f_c_2 = open(INPUT_ROOT_DIR + "seed02//control//best_last_org.txt", "r")
txt_c_2 = f_c_2.read()
size_c_2 = len(txt_c_2)
f_c_2.close()

# Seed03
f_c_3 = open(INPUT_ROOT_DIR + "seed03//control//best_last_org.txt", "r")
txt_c_3 = f_c_3.read()
size_c_3 = len(txt_c_3)
f_c_3.close()

# Seed04
f_c_4 = open(INPUT_ROOT_DIR + "seed04//control//best_last_org.txt", "r")
txt_c_4 = f_c_4.read()
size_c_4 = len(txt_c_4)
f_c_4.close()

# Seed05
f_c_5 = open(INPUT_ROOT_DIR + "seed05//control//best_last_org.txt", "r")
txt_c_5 = f_c_5.read()
size_c_5 = len(txt_c_5)
f_c_5.close()
```

Now that we have the sizes of all of the best organisms, we can find the average:

```python
# compute the average size
avg_c = (size_c_1 + size_c_2 + size_c_3 + size_c_4 + size_c_5) / 5
print("The average size of the genome for the control condition across all seeds is", avg_c, "base pairs.")
```
